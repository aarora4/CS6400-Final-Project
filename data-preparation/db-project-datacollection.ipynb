```python
cast_df = pd.DataFrame(columns=['movie_id', 'cast_ids', 'director_id'])
def get_cast_crew(movie_id):
    
    # https://api.themoviedb.org/3/movie/100/credits?api_key=b996cb6bd7c36029c4e2ee24a467d91b&language=en-US
    
    target_url = 'https://api.themoviedb.org/3/movie/{}/credits?api_key={}&language=en-US'.format(movie_id, api_key)
    
    try:
    
        with urlopen(target_url) as response:
            
            response_dict = json.loads(response.read())
            
            cast = response_dict['cast']
            crew = response_dict['crew']
            
            cast_ids = [x['id'] for x in cast]
            # crew_ids = [x['id'] for x in crew]
            
            director = 0
            
            for crewmate in crew:
                
                if crewmate['job'] == 'Director':
                    director = crewmate['id']
                    break
            
            cast_ids = np.unique(cast_ids)
            # crew_ids = np.unique(crew_ids)
            
            
            
            # print(cast_ids)
            # print(crew_ids)
            
            # print(director)
            
            cast_df.loc[len(cast_df)] = [movie_id, cast_ids, director]
           
        
    except:
        
        print("Couldn't get movie with id {}".format(movie_id))
        
for movie_id in range(100, 1000):
    get_cast_crew(movie_id)
```

```python
all_df = movie_df.merge(cast_df, on='movie_id')

print(all_df.head())
```

```python
movie_keywords_df = pd.DataFrame(columns=['movie_id', 'keywords'])

def get_keywords(movie_id):
    
    # https://api.themoviedb.org/3/movie/100/credits?api_key
    
    target_url = 'https://api.themoviedb.org/3/movie/{}/keywords?api_key={}&language=en-US'.format(movie_id, api_key)
    
    try:
        
        with urlopen(target_url) as response:
            
            response_dict = json.loads(response.read())
            
            keywords = response_dict['keywords']
            
            keywords = [x['id'] for x in keywords]
            
            movie_keywords_df.loc[len(movie_keywords_df)] = [movie_id, keywords]
            
    except:
        
        print("Couldn't get keywords for movie with id {}".format(movie_id))
        
for movie_id in range(100, 1000):
    get_keywords(movie_id)
```

```python
# perform tf-idf on movie overview to extract more keywords

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')

all_df['overview'] = all_df['overview'].fillna('')

tfidf_matrix = tfidf.fit_transform(all_df['overview'])

tfidf_matrix.shape

print(all_df['overview'][0])

doc = 0
feature_index = tfidf_matrix[doc,:].nonzero()[1]
tfidf_scores = zip(feature_index, [tfidf_matrix[doc, x] for x in feature_index])

# print the top 10 keywords by tf-idf score
keywords = []
for w, s in [(tfidf.get_feature_names()[i], s) for (i, s) in tfidf_scores]:
    keywords.append((w, round(s, 3)))
    # print(w, s)

keywords = sorted(keywords, key=lambda x: x[1], reverse=True)

print(keywords[:10])
    


from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

indices = pd.Series(all_df.index, index=all_df['original_title']).drop_duplicates()


def get_additional_keyword_recommendations(title, cosine_sim=cosine_sim):
    
    idx = indices[title]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    sim_scores = sim_scores[1:11]
    
    movie_indices = [i[0] for i in sim_scores]
    
    return all_df['original_title'].iloc[movie_indices], sim_scores

get_additional_keyword_recommendations('The Dark Knight')

# for movie_id in range(100, 1000):
#     get_additional_keyword_recommendations(all_df[all_df['movie_id'] == movie_id]['original_title'].iloc[0])
```

```python
edges_df = pd.DataFrame(columns=['source', 'target', 'type'])

def add_same_director_edges():
    
    for index, row in all_df.iterrows():
        
        if row['director_id'] != 0:
            
            director = row['director_id']
            
            all_refs = all_df[all_df['director_id'] == director]
            
            for ref in all_refs['movie_id']:
                
                if row['movie_id'] != ref:
                
                    edges_df.loc[len(edges_df)] = [row['movie_id'], ref, 'same_director']
                

def add_same_production_edges():
    
    for index, row in all_df.iterrows():
        
        production_companies = row['production_companies']
        
        # get all movies that have at least one of the same production companies
        
        all_refs = all_df[all_df['production_companies'].apply(lambda x: len(set(x).intersection(production_companies)) > 0)]
        
        for ref in all_refs['movie_id']:
            
            if row['movie_id'] != ref:
            
                edges_df.loc[len(edges_df)] = [row['movie_id'], ref, 'same_production']

# add_same_production_edges()

# print(edges_df.head())
# print(edges_df.shape)

def add_same_cast_edges():
    
    for index, row in all_df.iterrows():
        
        cast = row['cast_ids']
        
        # get all movies that have at least one of the same cast members
        
        all_refs = all_df[all_df['cast_ids'].apply(lambda x: len(set(x).intersection(cast)) > 0)]
        
        for ref in all_refs['movie_id']:
            
            if row['movie_id'] != ref:
            
                edges_df.loc[len(edges_df)] = [row['movie_id'], ref, 'same_cast']


# def add_similar_movie_edges():
    
#     for index, row in all_df.iterrows():
        
#         similar_movies = row['similar_movie_ids']
        
#         for ref in similar_movies:
            
#             if row['movie_id'] != ref and ref in all_df['movie_id'].values:
            
#                 edges_df.loc[len(edges_df)] = [row['movie_id'], ref, 'similar_movie']
                
def add_same_keyword_edges():
    
    for index, row in all_df.iterrows():
        
        keywords = row['keywords']
        
        # get all movies that have at least one of the same keywords
        
        all_refs = all_df[all_df['keywords'].apply(lambda x: len(set(x).intersection(keywords)) > 0)]
        
        for ref in all_refs['movie_id']:
            
            if row['movie_id'] != ref:
            
                edges_df.loc[len(edges_df)] = [row['movie_id'], ref, 'same_keyword']


add_same_director_edges()
add_same_production_edges()
add_same_cast_edges()
# add_similar_movie_edges()
add_same_keyword_edges()

print(edges_df.head())
print(edges_df.shape)
```

```python
def get_movie_recommendations(movie_id):
    
    # get all edges that have the movie_id as the source
    
    edges = edges_df[edges_df['source'] == movie_id]
    
    # count the most common targets
    
    target_counts = edges['target'].value_counts()
    
    # get the top 10 most common targets
    
    top_targets = target_counts[:10]
    
    # get the movie titles for the top 10 targets
    
    top_targets = top_targets.index.values
    
    top_targets = all_df[all_df['movie_id'].isin(top_targets)]
    
    top_targets = top_targets['original_title']
    
    return top_targets

print(get_movie_recommendations(161))
```

```python
from neo4j import GraphDatabase

class HelloWorldExample:

    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def get_recommendations_by_title(self, title):
        
        with self.driver.session() as session:
            
            recommendations = session.run("MATCH (n)-[r]->(n2) WHERE n.title = $title RETURN n2, COUNT(r) AS count ORDER BY COUNT(r) DESC LIMIT 10", title=title)
            
            print(recommendations.data())
            return recommendations
    
    def get_recommendations_by_id(self, movie_id):
        
        with self.driver.session() as session:
            
            recommendations = session.run("MATCH (n)-[r]->(n2) WHERE n.movieId = $movie_id RETURN n2, COUNT(r) AS count ORDER BY COUNT(r) DESC LIMIT 10", movie_id=movie_id)
            
            print(recommendations.data())
            return recommendations



greeter = HelloWorldExample("bolt://localhost:7687", "neo4j", "CS6400")
# greeter.print_greeting("hello, world")
# greeter.get_recommendations("The Lord of the Rings: The Return of the King")

greeter.get_recommendations_by_id('300')

greeter.close()
```
